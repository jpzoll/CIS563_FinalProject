# K-Means Clustering
## Spiritual Accounts & Actor Accounts

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('data/mood_and_f_moods2.csv')
```

```python
df['mood'] = df['mood'].fillna(0)
df['follower_mood'] = df['follower_mood'].fillna(0)
```

```python
df
```

```python
plt.scatter(df['mood'], df['follower_mood'])
```

```python
kmeans = KMeans(n_clusters=2)
kmeans
```

```python
cluster_values = list(kmeans.fit_predict(df[['mood', 'follower_mood']]))
df['cluster'] = cluster_values
```

# Clustering Graph

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Mood Among Spirituality & Actor Accounts on Twitter ')
plt.xlabel('Account Mood')
plt.ylabel("Follower Mood (Avg.)")
#figure(figsize=(18,12))
colorings = ['green', 'red']

#ax.scatter(df.mood, df.follower_mood, color = [colorings[cluster] for cluster in df['cluster']])
user_list = list(df['user'].values)

for i, user in enumerate(user_list):
    ax.annotate(user, (df.mood[i], df.follower_mood[i]))
    
ax.scatter(df.mood, df.follower_mood, color = [colorings[cluster] for cluster in df['cluster']])
```
