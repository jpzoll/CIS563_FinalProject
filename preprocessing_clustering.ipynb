# Preprocessing Accounts

```python
df['screen_name'] = df['screen_name'].astype(str).str.replace('@', '')
```

```python
def is_valid_account(s):
    try:
        x = twitter.get_user(screen_name=s)
        return True
    except:
        return False
```
